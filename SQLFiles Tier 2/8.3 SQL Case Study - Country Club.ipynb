#### Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members!

```python
df = pd.read_sql_query('SELECT name, \
                       CASE WHEN memid = 0 THEN SUM(slots)*guestcost ELSE SUM(slots)*membercost \
                       END AS revenue \
                       FROM Facilities INNER JOIN Bookings ON Facilities.facid = Bookings.facid \
                       GROUP BY name \
                       HAVING revenue < 1000', engine)

df
```

#### Q11: Produce a report of members and who recommended them in alphabetic surname,firstname

```python
df = pd.read_sql_query('SELECT member.firstname, member.surname, rec.firstname AS RecommenderFirstName, rec.surname AS RecommenderSurname \
                        FROM Members rec \
                        INNER JOIN Members member ON rec.memid = member.recommendedby \
                        ORDER BY member.surname', engine)

df
```

#### Q12: Find the facilities with their usage by member, but not guests

```python
df = pd.read_sql_query('SELECT Bookings.facid, name, COUNT(CASE WHEN  Bookings.memid > 0 THEN 1 END) AS MemberBookings \
                        FROM Bookings \
                        LEFT JOIN Facilities USING(facid) \
                        GROUP BY Bookings.facid', engine)

df
```

#### Q13: Find the facilities usage by month, but not guests

```python
df = pd.read_sql_query("SELECT strftime('%m', starttime) as Month, name, COUNT(name) as 'MemberBookings' \
                            FROM Bookings \
                            LEFT JOIN Facilities USING(facid)\
                            where memid != 0 \
                            GROUP BY month, name;", engine)

df
```
